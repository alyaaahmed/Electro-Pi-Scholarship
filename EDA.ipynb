```python
df.drop_duplicates(inplace=True)
df
```

```python
#Missing Value
missing_count = df.isnull().sum()
value_count=df.isnull().count()
percentage_missing=round(missing_count/value_count)*100
percentage_missing
```

```python
#visualize the percentage of the missing value
missing_data=pd.DataFrame({"number of missing data" :missing_count, "percentage of missing data" :percentage_missing})
bar_chart=missing_data.plot(kind ="bar")
```

```python
#remove any columns that had 80% or more of its data as null values

missing_count= df.isna().sum()
all_null = missing_count/ len(df)
missing_features = all_null[all_null > 0.80].index
df.drop(missing_features, axis=1, inplace=True)
df
```

```python
 #Removing Rows with Null Values
df.dropna(axis=0,inplace = True)
df
```

 Handling Outliers  using IQR(Interquartile range) Method 

```python
numerical_cols = [df.select_dtypes([np.int64,np.float64]).columns]

for col in numerical_cols:
    p0=df[col].min()
    p100=df[col].max()
    q1=df[col].quantile(0.25)
    q2=df[col].quantile(0.5)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lc = q1 - 1.5*iqr
    uc = q3 + 1.5*iqr
    df[col].clip(upper=uc,inplace=True ,axis=1)
    df[col].plot(kind='box')

print( "IQR for column: " , col ,"\n", "p0 = " , p0 ,", p100 = " , p100 ,", lc = " , lc ,", uc = " , uc , '\n' )

    
```

# Analysis using Charts

```python
# numerical col
from pandas.api.types import is_numeric_dtype
from pandas.api.types import is_string_dtype
for col in df:
    plt.figure(col)
    plt.title(col)
    plt.grid()
    if is_numeric_dtype(df[col]):
        try:
         df[col].plot(kind="hist",color="y",bins=10)
         df[col].plot(kind="box")
        except:
            print("no numeric data to plot in",col)
        
    elif is_string_dtype(df[col]):
        try:
          df[col].value_counts()[:30].plot(kind="bar")
        except:
            print("no categrical data to plot in" ,col)
```
